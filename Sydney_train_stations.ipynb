```python
import numpy as np
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
stations = pd.read_csv('data/Stations_Sydney.csv')
```

```python
stations.head(10)
```

```python
lat=-33.87
lng=151.1
sydney_map = folium.Map(location=[lat, lng], zoom_start=12)
```

```python
sydney_map
```

```python
locations = folium.map.FeatureGroup()

for lat, lng,name, line,dis in zip(stations.latitude, stations.longitude, stations.Name, stations['Served by'], stations['Distance from Central']):
    locations.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=8, 
            color='red',
            fill_color='yellow',
            fill_opacity=0.7,
            popup = 'Line '+ line+ ' Dist to Central ' + str(dis),
            tooltip = name 
        )
    )
```

```python
sydney_map.add_child(locations)
```
